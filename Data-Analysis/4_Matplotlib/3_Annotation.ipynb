```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-5, 5, 1000)
Y = np.sinc(X)

plt.title('A simple marker exercise')
plt.xlabel('Array variables')
plt.ylabel('Random variables')
plt.text(-5, 0.5, 'Matplotlib')  # -5 is the x value and 0.5 is y value.

plt.plot(X, Y,
         color='r',
         marker='o',
         markersize=9,
         markevery=30,
         markerfacecolor='w',
         linewidth=3,
         markeredgecolor='b');
```

```python
def norm_pdf(X, mu, sigma):
    """
    Normal probability density function.
    """
    a = -(X - mu)**2 / (2 * sigma**2)
    b = np.sqrt(2 * np.pi * sigma**2)
    return np.exp(a) / b
```

You can pass a dictionary as an argument to `plt.text()` with all the style parameters for the bounding box.

```python
plt.plot(X, norm_pdf(X, 0, 0.75), color='r', linestyle='solid')
plt.plot(X, norm_pdf(X, 0, 0.50), color='g', linestyle='dashed')
plt.plot(X, norm_pdf(X, 0, 0.25), color='b', linestyle='dashdot')

# Style parameters for the text box:
design = {'facecolor': 'y',
          'edgecolor': 'g', 
          'boxstyle': 'round', # {'round', 'square'}
          'pad': 0.5}  # Padding between the text and the perimeter of the box.

plt.text(-4, 1.5, 'Matplot Lib', bbox = design)
plt.plot(X, Y, c='k')
```

### Alignment

As we saw, the text we plot is bound by a box.
This box is used to align the text relative to the coordinates passed to `pyplot.text()`.
We can control the alignment with the vertical (`va`) and horizontal (`ha`) alignment parameters.

The vertical-alignment options are:

* `center`: Relative to the center of the textbox.
* `top`: Relative to the upper side of the textbox.
* `bottom`: Relative to the lower side of the textbox.
* `baseline`: Relative to the text baseline.

The horizontal-alignment options are:

* `center`: Relative to the center of the textbox.
* `left`: Relative to the left side of the textbox.
* `right`: Relative to the right side of the textbox.

### Arrows

```python
X = np.linspace(-5, 5, 1000)
Y = .25 * (X + 4) * (X + 1) * (X - 2)

plt.annotate('Biiiig Data',
             ha='center',
             va='bottom',
             xytext=(0, 10),
             xy=(1, -1), 
             arrowprops={'facecolor': 'orange',  # Arrow properties
                         'edgecolor': 'green',
                         'shrink':.05})

plt.plot(X, Y)
```

### Legends

Legend properties:

* `loc`: This is the location of the legend. The default value is `best`, which will place it automatically. Other options are `upper left`, `lower left`, `lower right`, `right`, `center left`, `center right`, `lower center`, `upper center` and `center`.
* `shadow`: Can be `True` or `False`, and it renders the legend with a shadow effect.
* `fancybox`: Can be `True` or `False` and renders the legend with a rounded box.
* `title`: Renders the legend with the title passed as a parameter.
* `ncol`: Indicates the number of columns for the legend.

```python
x = np.linspace(0, 6,1024)
y1 = np.sin(x)
y2 = np.cos(x)

plt.xlabel('Sin Wave')
plt.ylabel('Cos Wave')
plt.plot(x, y1, c='b', lw=3, label='Sin(x)')  # Labels can also be specified as an argument.
plt.plot(x, y2, c='r', lw=3, label='Cos(x)', ls='--')

# Add grid-lines to the figure:
plt.grid(True, lw=2, ls='--', c='.75')

plt.legend(loc='best',
           shadow=True,
           fancybox=True,
           title='Waves',
           ncol=1)
```

### Shapes

Paths for several kinds of shapes are available in the `matplotlib.patches` module.

```python
import matplotlib.patches as patches
```

```python
p_1 = patches.Circle((0, 0),  # (x coord, y coord)
                     radius=1,
                     color='.75')

p_2 = patches.Rectangle((2.5, -.5),
                        2, 1,
                        color ='.75')

p_3 = patches.Ellipse((0, -2.0),
                      2, 1,  # width, height
                      angle=45,
                      color='0')

# FancyBboxPatch := {'larrow', 'rarrow', 'round', 'round4', 'roundtooth', 'sawtooth', 'square'}
p_4 = patches.FancyBboxPatch((2.5, -2.5),
                             2, 1,
                             boxstyle='roundtooth',
                             color ='g')
plt.gca().add_patch(p_1)
plt.gca().add_patch(p_2)
plt.gca().add_patch(p_3)
plt.gca().add_patch(p_4)

plt.grid(True)
plt.axis('scaled')
```

`matplotlib.patches.Polygon()` takes as input a list of coordinates and interprets them as the vertices of a polygon.

```python
# 8 points linearly-spaced on a full circumference:
theta = np.linspace(0, 2 * np.pi, 8)
vertical = np.vstack( (np.cos(theta), np.sin(theta)) ).T
print(vertical)

plt.gca().add_patch(patches.Polygon(vertical, color ='y'))
plt.axis('scaled')
plt.grid(True)
```

The polygon can be embedded into a circle.

```python
theta = np.linspace(0, 2 * np.pi, 6)
vertical = np.vstack( (np.cos(theta), np.sin(theta)) ).T
print(vertical)

plt.gca().add_patch(plt.Circle((0, 0), radius=1.0, color ='b'))
plt.gca().add_patch(plt.Polygon(vertical, fill=None, lw=4.0, ls='dashed', edgecolor ='w'))
plt.axis('scaled')
plt.grid(True)
```

### Ticks

Ticks are the small marks on the axes of a figure.

```python
import matplotlib.ticker as ticker
```

```python
X = np.linspace(-20, 20, 1000)
Y = .25 * (X + 4) * (X + 1) * (X - 2)
ax = plt.axes()  # The object that manages the axes of the figure.

ax.xaxis.set_minor_locator( ticker.MultipleLocator(1) )
ax.xaxis.set_major_locator( ticker.MultipleLocator(5) )

plt.plot(X, Y, c='y')
plt.grid(True, which='major')  # which := {'minor', 'major', 'both'}
```

```python
name_list = ('Johannes', 'Serguey', 'Pedro', 'Zhou', 'Abidin')
value_list = np.random.randint(0, 99, size=len(name_list))
pos_list = np.arange(len(name_list))

ax = plt.axes()
ax.xaxis.set_major_locator( ticker.FixedLocator(pos_list) )
ax.xaxis.set_major_formatter( ticker.FixedFormatter(name_list) )

plt.bar(pos_list, value_list, color='m', align='center')
```
