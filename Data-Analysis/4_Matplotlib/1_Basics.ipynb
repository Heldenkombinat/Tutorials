### Plotting

```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np 
```

```python
x = list(range(100))
print(x)
plt.plot(x);
```

```python
y = [value**2 for value in x] 
print(y)
plt.plot(y);
```

```python
plt.plot(x, y)
```

### Using Numpy

Now let's see how we can use `np.linspace()`. You can find the complete documentation [here](http://docs.scipy.org/doc/numpy/reference/generated/numpy.linspace.html).

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

print(x)
plt.plot(x,y)
```

```python
x = np.linspace(-3, 2, 200)
y = x**2 - 2*x + 1
plt.plot(x, y)
```

# Multiple plots per figure

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
z = np.cos(x)

# Matplotlib assigns different colors to each plot:
plt.plot(x, y) 
plt.plot(x, z)
```

```python
data = np.loadtxt('numbers.txt')
print(data)

x = data[:, 0]
y = data[:, 1]
plt.plot(x, y)
```

```python
data = np.loadtxt('numbers.csv', delimiter=',')
print(data.T)

for sequence in data.T:
    plt.plot(sequence)
```

### Bar charts

```python
x = [5, 10 ,15, 20, 25]
y = [1, 2, 3, 4, 5]
plt.bar(x, y)
```

You can change the thickness of a bar with the argument `width`. By default the bar will have a width of 0.8 units.

```python
plt.bar(x, y, width=4)
```

`barh()` is a graph with horizontal bars.

```python
plt.barh(x, y)
```

### Multiple bar charts

```python
x = np.arange(4) 

Y = [[1, 11, 51, 21],
     [2, 12, 52, 22],
     [3, 13, 53, 23]]

plt.bar(x-0.25, Y[0], color='r', width=0.25)
plt.bar(x+0.00, Y[1], color='g', width=0.25)
plt.bar(x+0.25, Y[2], color='b', width=0.25)
```

Stacked bar charts:

```python
x = range(4)
p = [5, 30, 45, 22]
q = [5, 25, 50, 20]

plt.bar(x, p, color ='b')
plt.bar(x, q, color ='y', bottom=p)
```

Plotting more than 2 values:

```python
X = np.arange(4)
A = np.array([5, 10, 15, 20])
B = np.array([10, 20, 40, 80])
C = np.array([10, 10, 10, 10])

plt.bar(X, A, color='r')
plt.bar(X, B, color='g', bottom=A)
plt.bar(X, C, color='b', bottom=A+B)
```

```python
x = np.arange(4)
left = np.array([5, 10, 15, 20]) 
right = np.array([5, 10, 20, 40])

plt.barh(x, -left, color ='r')  # We can use minus to create back-to-back bars.
plt.barh(x, right, color ='g')
```

* The red bar is the median of the distribution

```python
d = np.random.randn(100)
plt.boxplot(d)
```

Multiple box plots:
