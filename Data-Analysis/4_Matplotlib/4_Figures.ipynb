# Matplotlib - Figures

This tutorial is based on the open-source [notebooks](https://github.com/manujeevanprakash/Matplot-lib-Basics) of Manu Jeevan Prakash.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
T = np.linspace(-np.pi, np.pi, 1000)

fig, (ax0, ax1) = plt.subplots(ncols=2)
ax0.plot(np.sin(2 * T), np.cos(0.5 * T), c='k')
ax1.plot(np.cos(3 * T), np.sin(T), c='k')
```

### Aspect ratio

```python
T = np.linspace(0, 2 * np.pi, 1000)

fig, ax = plt.subplots(ncols=1)
ax.plot(2 * np.cos(T), np.sin(T), c='k', lw=3)

ax.set_aspect('equal')  # Remove this line of code and see how the figure looks.
```

```python
X = np.linspace(-10, 10, 1000)
Y1, Y2 = np.sinc(X), np.cos(X)

plt.figure(figsize=(15, 5))  # Sets size of the figure.

plt.plot(X, Y1, c='r', lw=3)
plt.plot(X, Y2, c='.75', lw=3)
```

```python
X = np.linspace(-5, 5, 1000)
plt.plot(X, np.sinc(X), c='k')

plt.ylim(-.5, 1.5)  # Sets the limits on the vertical axis.
```

```python
X = np.linspace(-5, 5, 1000)
Y = np.sinc(X)

# Coordinates of subplot:
X_sub = np.linspace(-3, 3, 1000)
Y_sub = np.sinc(X_sub)

plt.plot(X, Y, c='b')

# Coordinates, length and width of the subplot frame:
sub_axes = plt.axes([.65, .55, .2, .2])

sub_axes.plot(X_sub, Y_sub, c='r')
```

### Log scale

With the function `plt.xscale()` and `plt.yscale()` we can set the axis scales to be logarithmic.
The logarithm base is 10 by default, but it can be changed with the optional parameters `basex` and `basey`.

```python
X = np.linspace(1, 10, 1000)

plt.yscale('log')
plt.plot(X,         X, c='.00', lw=2, label=r'$f(x)=x$', ls='--')
plt.plot(X,     10**X, c='.75', lw=2, label=r'$f(x)=e^x$')
plt.plot(X, np.log(X), c='.75', lw=2, label=r'$f(x)=\log(x)$')
plt.legend()
plt.show()
```

### Polar Coordinates

```python
T = np.linspace(0, 2 * np.pi, 1024)
plt.axes(polar=True) # show polar coordinates
plt.plot(T, 1 + .25 * np.sin(16 * T), c='k')
plt.show()
```

Making use of the `patches` module from matplotlib.

```python
ax = plt.axes(polar=True)
theta = np.linspace(0, 2 * np.pi, 8, endpoint=False)
radius = .25 + .75 * np.random.random(size=len(theta))
points = np.vstack((theta, radius)).transpose()
plt.gca().add_patch(patches.Polygon(points, color='.75'))
plt.show()
```

Create a file named `sine.png`. Its resolution will be 800 x 600 pixels, in 8-bit colors (24-bits per pixel).

```python
x = np.linspace(-6, 6, 1024)
y = np.sin(x)
plt.plot(x, y)
plt.savefig('sine.png', c='y', transparent=True)  # Writes that data to a file.
```

`dpi` is dots per inch. 300*8 x 6*300 = 2400 x 1800 pixels.

```python
theta = np.linspace(0, 2 * np.pi, 8)
points = np.vstack((np.cos(theta), np.sin(theta))).T
plt.figure(figsize=(6.0, 6.0))
plt.gca().add_patch(plt.Polygon(points, color ='r'))
plt.axis('scaled')
plt.grid(True)
plt.savefig('polar.png', dpi=300)  # Try out 'polar.pdf', polar.svg'
```
