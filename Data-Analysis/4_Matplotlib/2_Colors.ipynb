### Color values

The basic built-in colors in Matplotlib can be specified with a single letter:

* `b`: blue
* `g`: green
* `r`: red
* `c`: cyan
* `m`: magenta
* `y`: yellow
* `k`: black
* `w`: white

#### Scatterplots

Colors go from black (`0.0`) to white (`1.0`).

```python
a = np.random.standard_normal((100, 2))
a += np.array((1, 1))  # Center the distribution at (1, 1)

b = np.random.standard_normal((100, 2))
b += np.array((-1, -1))  # Center the distribution at (-1, -1)

plt.scatter(a[:, 0], a[:, 1], color='.2')
plt.scatter(b[:, 0], b[:, 1], color='.8')
```

We can also specify the color of the edge.

```python
c = np.random.standard_normal((100, 2))

plt.scatter(c[:, 0], c[:, 1], color='1.0', edgecolor='0.0')
```

#### Bar charts

Plot a sequence of bars, where each has one of four different colors.

```python
values = np.random.randint(100, size=50)

color_set = ['.00', '.25', '.50','.75']
color_list = np.random.choice(color_set, len(values))

plt.bar(np.arange(len(values)), values, color=color_list)
```

#### Pie charts

`plt.pie()` will cycle through the colors when provided with less colors than values.
For example, if we have eight values and four colors, each color will be used twice.

```python
hi = np.random.randint(10, size=8)
color_set =['.00', '.25', '.50', '.75']

plt.pie(hi, colors=color_set);
```

#### Box plots

```python
color_set = ['r', 'g', 'b']
values = np.random.randn(100)
figure = plt.boxplot(values)

for key, values in figure.items():
    
    color = np.random.choice(color_set)
    print('{}: {}'.format(key, color))
    
    for v in values:
        v.set_color(color)
```

### Color Maps

The `matplotib.cm` module provides functions to create and use colormaps.
It also provides a comprehensive range of predefined color maps.

```python
import matplotlib.cm as cm
```

To start, we will use **Hue, Saturation, Value** (HSV).
You can find information about the HSV color map in the following [documentation](https://en.wikipedia.org/wiki/HSL_and_HSV).

```python
N = 300
angle = np.linspace(0, 10 * 2 * np.pi, N)
radius = np.linspace(0.5, 1.0, N)

X = radius * np.cos(angle)
Y = radius * np.sin(angle)

plt.scatter(X, Y, c=angle, cmap=cm.hsv)
```

We can create our own color maps as well.
For this, we need additional functions from `matplotlib.colors`.

```python
import matplotlib.colors as cl

# Normalizing object which scales the data between the given limits:
norm = cl.Normalize(0, 100)

# The colormap used to map normalized data values to RGBA colors:
cmap = cm.binary

# This is a mixin class to support scalar data to RGBA mapping:
color_converter = cm.ScalarMappable(norm, cmap)

vals = np.random.randint(100, size=50)
n_vals = np.arange(len(vals))
custom_cmap = color_converter.to_rgba(vals)

plt.bar(n_vals, vals, color=custom_cmap)
```

### Line Styles

```python
def norm_pdf(X, mu, sigma):
    """
    Normal probability density function.
    """
    a = -(X - mu)**2 / (2 * sigma**2)
    b = np.sqrt(2 * np.pi * sigma**2)
    return np.exp(a) / b
```

```python
help(norm_pdf)
```

```python
data = np.linspace(-1, 1, 10000)

plt.plot(data, norm_pdf(data, 0, 0.75), color='r', linestyle='solid')
plt.plot(data, norm_pdf(data, 0, 0.50), color='g', linestyle='dashed')
plt.plot(data, norm_pdf(data, 0, 0.25), color='b', linestyle='dashdot')
```

```python
N = 5
A = np.random.random(N)
B = np.random.random(N)
X = np.arange(N)

plt.bar(X, A, color='.75')
plt.bar(X, B, color='k', linestyle ='dashed', bottom=A)
```

```python
X = np.linspace(-3, 3, 10000)

for i in range(100):
    samples = np.random.standard_normal(100)
    mu, sigma = np.mean(samples), np.std(samples)
    plt.plot(X, norm_pdf(X, mu, sigma), color = '.75', linewidth=.5)

plt.plot(X, norm_pdf(X, 0, 1), color ='0', linewidth=3)
```

### Fill surfaces with pattern

Some markers or `hatch` attributes that we can use are:

`/`, `\`, `|`, `-`, `+`, `x`, `o`, `O`, `.` and `*`.

```python
N = 5
A = np.random.random(N)
B = np.random.random(N)
X = np.arange(N)

plt.bar(X, A, color='w', hatch='*')
plt.bar(X, B, color='m', hatch='O', bottom=A)
```

### Marker styles

```python
from IPython.display import Image, display

a = Image(filename='unfilled-markers.png')
b = Image(filename='filled-markers.png')

display(a, b)
```

When you have dense data, you can place marks every N elements.

```python
X = np.linspace(-5, 5, 1000)
Ya = np.sinc(X)
Yb = np.sinc(X) + 1

plt.plot(X, Ya, marker='o', color='.75', markevery=20)
plt.plot(X, Yb, marker='^', color='.00', markevery=20)
```

With the attribute `s` we can specify the marker size.

```python
A = np.random.standard_normal((100, 2))
A += np.array((-1, -1))

B = np.random.standard_normal((100, 2))
B += np.array((1, 1))

plt.scatter(A[:, 0], A[:, 1], color='c', s=100)
plt.scatter(B[:, 0], B[:, 1], color='m', s=25)
```

```python
# more about markers
X = np.linspace(-5, 5, 1000)
Y = np.sinc(X)

plt.plot(X, Y,
         color           = 'r',
         linewidth       = 3,
         marker          = 'o',
         markerfacecolor = 'w',
         markeredgecolor = 'b',
         markersize      = 10,
         markevery       = 20)
```
